```python
from __future__ import division, print_function
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# I will revisit some small groups later, after looking at more global properties.
# I will make sure of the currency info and separate currency amounts
# otherwise we will mix up currencies when handling
# transaction amounts and perk amounts.
# Easiest thing to start is just do USD to begin with.
dataUSD = data[ data.currency == 'USD']
dataUSD.describe()
```

```python
# This is a good choice as it comprises ~88% of all data anyway
N = len(dataUSD)
print(N)
```

```python
# Looking at perk_id and perk_amount again, is there any difference
# in the presence of a claimed perk?
# Note from the above summary that the max is huge and the min is tiny...
# Just how many are above 1 std dev?
avg = dataUSD.transaction_amount.mean()
std = dataUSD.transaction_amount.std()
print('Percent Below One std ', len(dataUSD[ dataUSD.transaction_amount < avg + std ]) / N*100 )
print('Percent Below Two std ',len(dataUSD[ dataUSD.transaction_amount < avg + 2*std ]) / N*100 )
print('Two std\'s: ', avg + 2* std)
```

```python
# As almost 99% of the data are within the first 2 std's
# I will cut off the histogram a bit beyond this value (at 600)
print('Histogram contains {:.2%} of the data.'.format(len(dataUSD[ dataUSD.transaction_amount < (avg + 600) ]) / N ))

extra_params = plt.hist(dataUSD.transaction_amount[ dataUSD.transaction_amount < 600 ].values,bins = np.arange(0,600,10), color = '#cccccc')
plt.xlabel('Transaction Amount')
plt.ylabel('Number of Transactions')
plt.title("All Data")
```

```python
# I can play around with the bin width a bit, but one emerging
# pattern is the spikes at "round numbers", i.e, 50,100,200...
# Is that from 1) The perk structure or 2) Human behavior?
# Split into perks and no perks data sets and replot the histograms
with_perks = dataUSD[np.isnan(dataUSD.perk_id) == False]
wout_perks = dataUSD[np.isnan(dataUSD.perk_id) == True]
```

```python
# First, what is the distribution of the perk amount?
extra_params = plt.hist(with_perks.perk_amount[ with_perks.transaction_amount < 600 ].values,bins = np.arange(0,600,10))
plt.ylabel('Number of Transactions')
plt.xlabel('Perk Amount')
plt.title("With Perk Data")
```

```python
# Well, some strong structure without any claimed perk as well.
# We might not be able to say much at this point about how strongly
# a set perk drives user contributions.

# Let's try to answer something more economical, the relative
# money contributed per contribution for perks and no perks.
# First, a summary of data with a claimed perk.
print("Claimed perk summary")
with_perks.describe()
```

```python
# Without a claimed perk
print("No claimed perk summary")
wout_perks.describe()
```

```python
# First I note that both distrubtion means are similar (~78 and ~80).
# For the std there is a much larger value when no perk was claimed

# Look at the quartiles in the two tables. Comparing quartiles across tables, the with_perks quartile is larger.
# e.g. for the 25%, the transacted amount is 20USD for with_perks and 10USD for wout_perks.
# Let's get the total amounts that fall below each quartile.
def sum_quantile(q, column):
    quantile = column.quantile(q)
    return column[ column <= quantile ].sum()

def count_quantile(q, column):
    quantile = column.quantile(q)
    return column[ column <= quantile ].count()
```

```python
# A lot of code, see below output...
N_perk = len(with_perks.transaction_amount)
N_wout = len(wout_perks.transaction_amount)
perk_tot = sum_quantile(1, with_perks.transaction_amount)
wout_tot = sum_quantile(1, wout_perks.transaction_amount)
gains = []
print("\t\t\t\tPerk\t\tNo Perk")
for q in [0.25, 0.5, .75, 1]:
    print("up to q {0:.0%}".format(q))
    
    perk_sum = sum_quantile(q, with_perks.transaction_amount)
    wout_sum = sum_quantile(q, wout_perks.transaction_amount)
    
    perk_count = count_quantile(q, with_perks.transaction_amount)
    wout_count = count_quantile(q, wout_perks.transaction_amount)
    
    per_perk = perk_sum / perk_count
    per_wout = wout_sum / wout_count
    print("\tTotal (USD)\t\t{0}\t\t{1}".format(perk_sum,wout_sum))
    print("\tPer Contribution\t{0:.2f}\t\t{1:.2f}".format(per_perk, per_wout))
    print("\tPercent Total\t\t{0:.1%}\t\t{1:.1%}".format(perk_sum/perk_tot, wout_sum/wout_tot))
    
    conv_gain = ( per_perk - per_wout) * wout_count
    rel_gain = conv_gain / (perk_tot + wout_tot)
    gains.append((conv_gain,rel_gain))
    
```

```python
# This seems actionable.
# Regarding all contributions with a claimed perk, q25 contains 4.4% of all contributions in USD.
# In the case of contributions without a perk, q25 contains only 2.2% of all contributions in USD.
# We would like if the lower 25% of contributions (although small) made up as large a slice
# of the total pie as possible.
# Imagine if all the contributions without a claimed perk behaved like those with claimed perks.
# The table below shoes the percent increase in overall contributions if the
# "Per Contribution" value above changed from the "No Perk" value to the "Perk" value.
# e.g. in the 25%, from 6.67 USD to 12.21 USD.
print("quantile\tcontribution increase\tpercent of all contribution increase")
for q, each in zip([.25,.5,.75,1],gains):
    print("\tq={0:.0%}\t{1:.2f} USD \t\t{2:.2%}".format(q,each[0], each[1]))
```

```python
# Another intesting thing to look at is peak usage.
# this could be relevant for marketing, when to target ads / when people are using the service
# the most. The latter has additional application in engineering with regard to server load.
from collections import defaultdict
by_hour = defaultdict(float)
by_weekday = defaultdict(float)
by_date = defaultdict(float)

for timestamp, amount in zip(data.transacted_at, data.transaction_amount):
    hour = timestamp.hour
    day = timestamp.weekday()
    date = timestamp.date()
    by_hour[hour] += amount
    by_weekday[day] += amount
    by_date[date] += amount
```

```python
# Let's make some plots with date/time as the x axis showing how
# contribution varies with time.
import matplotlib.dates as mdates
def plot_series(by_dict):
    x1 = by_dict.keys()
    y1 = by_dict.values()
    z = sorted(zip(x1,y1))
    x, y = zip(*z)
    fig, ax = plt.subplots(1)
    
    fig.autofmt_xdate()
    ax.plot(x,y,)

# First let's do time of day.
# Using the aggregation "by_hour", we can
# see what time of day people are using the service
# and making contributions.
plot_series(by_hour)
plt.xlabel("Hour of a 24 hour day")
plt.ylabel("Total Contribution Amounts")
plt.savefig("Hours.pdf")
```

```python
# Above, there are more sales between 7am and 6pm PST
# This makes sense given that most transactions were in USD.

plot_series(by_weekday)
plt.xticks(by_weekday.keys(), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.ylabel("Total Contribution Amounts")
plt.savefig("DayOfWeek.pdf")
```

```python
# Above notice the peak on Friday

# Now let's do the overall series treating each date of the
# month as a bin for contributions.
plot_series(by_date)
plt.ylabel("Total Contribution Amounts")
```

```python
# Above the total amount per day is shown over the course of about one month
# It is hard to see a trend, which is unexpected because there is a clear trend 
# when looking at which days of the week are favored.

# Let's consider just the number of transactions rather than the summed amount
# of the transaction and plot again.
by_date_vol = defaultdict(float)
for timestamp, amount in zip(data.transacted_at, data.transaction_amount):
    date = timestamp.date()
    by_date_vol[date] += 1
```

```python
# Easier to see when plotting number of transactions.
# Now there is the same clear, weekly seasonal behavior.
plot_series(by_date_vol)

plt.ylabel("Number of Contributions")
```

```python
grouped = dataUSD.groupby('contributor_country')
```

```python
subsets = []
for cc, subset in grouped:
    # We only care if there are enough entries
    # to say something meaningful, lets go with 25...
    if len(subset) >= 25:
        print(cc, len(subset))
        print(subset.transaction_amount.mean())
        subsets.append((cc,subset))
```

```python
# 12 country names passed the test so let's try a grid and plot them...
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(12,8))
fig.autofmt_xdate()
for ax, ccsub in zip(axes.ravel(), subsets):
    cc = ccsub[0]
    subset = ccsub[1]
    by_vol = defaultdict(float)
    for timestamp, amount in zip(subset.transacted_at, subset.transaction_amount):
        date = timestamp.date()
        by_vol[date] += 1
    x1 = by_vol.keys()
    y1 = by_vol.values()
    z = sorted(zip(x1,y1))
    x, y = zip(*z)
    ax.plot(x,y)
    ax.set_title(cc)
```

```python
# Those are not the easiest to understand; let's try grouping foreign markets together.
foreign = data[data.contributor_country != 'United States']
for timestamp, amount in zip(foreign.transacted_at, foreign.transaction_amount):
    date = timestamp.date()
    by_vol[date] += 1
plot_series(by_vol)
```

```python
# And that brings us back to the same weekly peak structure we saw for the US markets.
# Let's take one final look at the mean and counts of contributions by country.
per_contribution = []
for cc, subset in grouped:
    mean = subset.transaction_amount.mean()
    count = subset.transaction_amount.count()
    if cc == 'United States':
        print(mean, count)
    if count > 3:
        per_contribution.append((cc,mean,count))
```

```python
l,x,y = zip(*per_contribution)
plt.scatter(x,y)
plt.ylim([0,300])
for label,x,y in per_contribution:
    plt.annotate(label, xy=(x,y), xytext=(20+(x%5)*30,20+(y%5)*30),
                textcoords = 'offset points', ha = 'right', va = 'bottom',
                bbox = dict(boxstyle = 'round, pad=0.5', fc = "#cccccc", alpha = .2),
                arrowprops=dict(arrowstyle = '->', connectionstyle = 'arc3, rad=0'))
    print(label)
plt.ylabel("Number of contributions")
plt.xlabel("Mean contribution")
```
