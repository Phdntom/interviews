```python
# Can't do much more without doing a regression model.
# Before that, let's clean up the data a bit more in the
# original dataframe.
intlist = []
for each in dfclean['OnAc?']:
    if each:
        intlist.append(1)
    else:
        intlist.append(0)
dfclean['OnAc?'] = intlist
dfclean.head()

```

```python
# On to our regression model use statsmodels.
# The reason is these are count processes bounded below by zero
# and unbounded above, so Poisson regression seems most fitting.
import statsmodels.api as sm
```

```python
# Let's treat each year grouping as a category.
df_2013 = dfclean[dfclean['Year'] == 2013]
df_2012 = dfclean[dfclean['Year'] == 2012]
df_2011 = dfclean[dfclean['Year'] == 2011]
df_2010 = dfclean[dfclean['Year'] == 2010]
```

```python
df_2010.head()
```

```python
# Here it is for 2013
keep_cols = ['Journal', 'OnAc?']
X_13 = sm.add_constant(df_2013[keep_cols])
y_13 = df_2013['Citation Count']
X_13.head()
```

```python
pois = sm.Poisson(y_13,X_13).fit()
```

```python
pois.summary()
```

```python
# As these coefficients are for a Poisson regression
# we can get the rate difference by exponentiating them.
# This means that for every unit change in the variable shown
# the multiplicative increase is shown.
np.exp(pois.params)
```

```python
# So increasing 'Journal' by 1 gives a 7% increase in citations.
# Increasing AC from 0 to 1 gives a 45% increase.
```

```python
# Here is a simple stacked procedure to carry out the above
# on any dataframe indexed out of dfclean by the 'Year' column.
def Poisson_Analysis(df):
    keep_cols = ['Journal', 'OnAc?']
    X = sm.add_constant(df[keep_cols])
    y = df['Citation Count']
    
    pois = sm.Poisson(y,X).fit(method='bfgs')
    
    print(pois.summary())
    print("\n")
    print("How unit change in parameters changes the citation rate?")
    print(np.exp(pois.params))
```

```python
# Check that we get the same results as before? Yep!
Poisson_Analysis(df_2013)
```

```python
# And the other years now.
Poisson_Analysis(df_2012)
```

```python
Poisson_Analysis(df_2011)
```

```python
Poisson_Analysis(df_2010)
```

```python
# Just for comparison we can do the entire dataframe
# not splitting on 'Year' differences.
Poisson_Analysis(dfclean.dropna())
```

```python
# So here we can conclude that while the Journal impact factor affects
# the citation rate a small amount, (varying depending on year selected),
# in all years examined, the affect of being on academia.edu is a very strong
# change in the citation rate in each year (33% to 46%).
```

```python
# Now in the last regression on the full data set, we have mixed categories
# i.e. data from different years, and we start to see some huge increases in the
# affect of the 'OnAc?' predictor variable. I suspect this is due to the
# exact type of confounding we tried so hard to avoid by splitting up the years.

# One interesting thing to do is use the whole data, rather than the year
# and use a continuous "time since publication" predictor in the same
# kind of analysis. Ideally we would see that when that is taken into account
# the impact of 'OnAc?' remains strong, even in the presence of the
# year of publication effects.
```
