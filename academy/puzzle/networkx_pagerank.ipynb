```python
from __future__ import division, print_function
import networkx as nx
```

```python
node_file = "academia-ri-nodes.txt"
edge_file = "academia-ri-edges.txt"
```

```python
# Each line is a node id and a topic name property of that node
node_list = []
with open(node_file,"r") as f:
    node_header = f.readline()
    node_contents = f.readlines()
    for node_line in node_contents:
        node_id, name = node_line.strip().split("\t")
        entry = (int(node_id), name)
        node_list.append(entry)
```

```python
# Each edge has a source/target node id and a weight
edge_list = []
with open(edge_file,"r") as f:
    edge_header = f.readline()
    edge_contents = f.readlines()
    for edge_line in edge_contents:
        source, target, weight = edge_line.strip().split("\t")
        entry = int(source), int(target), float(weight)
        edge_list.append(entry)
```

```python
def build_graph(node_list, edge_list):
    G = nx.Graph()
    for node_id, name in node_list:
        G.add_node(node_id, name=name)
    
    for src, tar, weight in edge_list:
        G.add_edge(src, tar, weight=weight, distance=1/weight)
        
    return G
```

```python
# Build up the Graph G from node and edge lists
# with name and weight properties respectively
G = build_graph(node_list, edge_list)
```

```python
# Straight indexing gives the edges of node [n], none right now
# .node indexing gives other keynamed properties of node [n]
print(G[8])
print(G.node[8])
```

```python
# Find the max degree by building a list of all degrees
# 2-tupled with the node name
degrees = []
for node in G.nodes_iter():
    degree = G.degree(node)
    degrees.append((G.node[node]['name'], degree))
```

```python
# Sort descending by the degree
degrees.sort(key=lambda x: x[1],reverse=True)
```

```python
# This allows a summary for a single node which
# does the lookups for the names of other nodes.
def explain_G(G, node_id):
    parent_node = G.node[node_id]['name']
    print("{0} is connected to these topics:\n".format(parent_node))
    print("\t{0:25}\t{1:10}".format('Field','Relatedness'))
    for n in G[node_id]:
        name = G.node[n]['name']
        weight = G[node_id][n]['weight']
        print("\t  {0:25}\t{1:10}".format(name,weight))
        
explain_G(G,8)
```

```python
# With the weighted graph built, we can
# find the page rank of all nodes.
rank = nx.pagerank(G)
# And can also do a "binary" page rank
# which treats all weights as unity.
unit_rank = nx.pagerank(G, weight=None)
```

```python
# build lists from the dictionaries
# to be sorted.
rank_list = []
unit_rank_list = []
for key in rank:
    rank_list.append((key,rank[key]))
    unit_rank_list.append((key,unit_rank[key]))
```

```python
# Sort descending
rank_list.sort(key=lambda x: x[1], reverse=True)
unit_rank_list.sort(key=lambda x: x[1], reverse=True)
```

```python
# Remap the node keys to their real names
names_rank = []
names_unit_rank = []

for key, score in rank_list:
    names_rank.append((G.node[key]["name"],score))
for key,score in unit_rank_list:
    names_unit_rank.append((G.node[key]["name"],score))
```
