```python
def logit_stack(p_func):
    
    x = np.linspace(0,5,200)
    y = np.random.uniform(size=200) < p_func(x)
    
    df = pd.DataFrame(zip(y,x))
    df.columns = ['y','x']
    df['intercept'] = 1.0
    
    train_cols = df.columns[1:]
    logit = sm.Logit(df['y'], df[train_cols])
    result = logit.fit()
    
    alpha, beta = result.params[0], result.params[1]
    ci = result.conf_int()
    alphaL, alphaU = ci[0][0], ci[1][0]
    betaL, betaU = ci[0][1], ci[1][1]

    yL = alphaL * x + betaL
    yU = alphaU * x + betaU   
    y_hat = alpha * x + beta
    
    p = 1 / ( 1 + exp(-y_hat))
    pU = 1 / ( 1 + exp(-yU))
    pL = 1 / ( 1 + exp(-yL))
    
    
    plt.plot(x, p, 'r--',linewidth=3)
    plt.plot(x, pL, 'g--',linewidth=3)
    plt.plot(x, pU, 'g--',linewidth=3)
    plt.scatter(x,y,color='k',marker='^')
    plt.scatter(x,p_func(x),color='c',marker='+')
    
logit_stack(random_sig)
```

```python
def flat(x,const=0.6):
    
    return np.array([const for i in x])
def linear(x,m=0.3):
    
    return np.array( [i*m for i in x ])
```

```python
logit_stack(flat)
```

```python
logit_stack(linear)
```
